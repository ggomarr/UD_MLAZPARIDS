```python
with open(work_dir+'Market_Basket_Optimisation.csv') as trans_file:
    trans_lst=trans_file.readlines()
len(trans_lst)
```

```python
all_products=set()
for trans_pos in range(len(trans_lst)):
    trans_lst[trans_pos]=set([ t.strip() for t in trans_lst[trans_pos].split(',') ])
    all_products.update(trans_lst[trans_pos])
trans_lst[:3]
```

```python
len(all_products)
```

# Finding the rules

```python
min_buys_per_day=3
min_support=min_buys_per_day*7/len(trans_lst)
min_support
```

```python
max_rule_order=3
min_support=0.01
rules={}
num_trans=len(trans_lst)

all_rules=[]
for rule_order in range(1,max_rule_order+1):
    if rule_order==1:
        potential_rules=[ ({product},) for product in all_products ]
    else:
        potential_rules=itertools.product(rules[rule_order-1],rules[1])
    rules[rule_order]=[]
    for rule in potential_rules:
        rule=set([inner for outer in rule for inner in outer])
        if len(rule)==rule_order and not rule in rules[rule_order]:
            rule_support=1.0*sum([rule.issubset(trans) for trans in trans_lst])/num_trans
            if rule_support>min_support:
                rules[rule_order].append(rule)
                all_rules.append((rule_order,rule_support,rule))
all_rules.sort(key=lambda x: (x[0],x[1]),reverse=True)
all_rules
```
