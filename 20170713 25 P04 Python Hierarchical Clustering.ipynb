# Dendogramming

```python
my_den=scipy.cluster.hierarchy.dendrogram(scipy.cluster.hierarchy.linkage(X_scaled,method='ward'))
plt.title('An awesome dendogram')
plt.xlabel('Customers')
plt.ylabel('Distance')
plt.show()
```

```python
def find_num_clusters(X,num=1,method='ward'):
    my_link=scipy.cluster.hierarchy.linkage(X,method=method)
    dist=[0]+[m[2] for m in my_link]
    deltas=[dist[i]-dist[i-1] for i in range(1,len(dist))]
    deltas_lst=sorted(deltas,reverse=True)[0:num]
    out=[]
    for delta in deltas_lst:
        pos=deltas.index(delta)
        out.append((len(X)-pos,delta,dist[pos],dist[pos+1]))
    return out

find_num_clusters(X_scaled,num=5)
```

# Hierarchical Clustering

```python
num_clusters_lst=range(1,10+1)
```

```python
for num_clusters in num_clusters_lst:
    print('=== {:2d} clusters ==='.format(num_clusters))
    clusterer=sklearn.cluster.AgglomerativeClustering(n_clusters=num_clusters,affinity='euclidean',linkage='ward')
    cluster=clusterer.fit_predict(X_scaled)
    color_lst=plt.cm.get_cmap('jet',num_clusters)
    for c in range(num_clusters):
        plt.scatter(X_scaled[cluster==c,0],X_scaled[cluster==c,1],
                    color=color_lst(c),label='Cluster {:02d}'.format(c+1))
    plt.legend()
    plt.show()
```
