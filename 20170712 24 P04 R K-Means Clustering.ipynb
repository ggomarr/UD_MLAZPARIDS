# Scale

```r
df=scale(df)
head(df)
```

# K-Means

```r
set.seed(6)
num_clusters=1:10
```

```r
model_lst=list()
for(num in num_clusters){
    model_lst[num]=sum(kmeans(df,num)$withinss)
}
plot(num_clusters,model_lst,'b',main='Here is your E.L.B.O.W.!',xlab='Number of clusters',ylab='WCSS')
```

```r
clusters=1:5
```

```r
set.seed(29)
for(n in clusters){
    cluster=kmeans(df,n,iter.max=300,nstart=10)
    clusplot(df,cluster$cluster,lines=0,shade=TRUE,color=TRUE,span=TRUE,plotchar=FALSE,
             main='Our clients',xlab='Annual Income',ylab='Spending Score',sub='')    
}
```
