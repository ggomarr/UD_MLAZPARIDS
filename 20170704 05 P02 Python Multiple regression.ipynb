# Feature selection using statsmodels.formula.api.OLS

## Adding an explicit 'intercept feature'

```python
X_train_ext=np.append(np.ones((X_train.shape[0],1)),X_train,1)
X_test_ext=np.append(np.ones((X_test.shape[0],1)),X_test,1)
```

## Backward elimination

```python
p_limit=0.05
be_feat=list(range(X_train_ext.shape[1]))
done=False
while not done:
    be_mult_reg_OLS=sm.OLS(Y_train,X_train_ext[:,be_feat]).fit()
    p_max=max(be_mult_reg_OLS.pvalues)
    if p_max>p_limit:
        p_max_pos=np.argmax(be_mult_reg_OLS.pvalues)
        print('Removing feature {} with p {:.3f}'.format(be_feat[p_max_pos],p_max))
        be_feat=be_feat[:p_max_pos]+be_feat[p_max_pos+1:]
    else:
        done=True
be_mult_reg_OLS.summary()
```

## Forward selection

```python
p_limit=0.05
fs_feat=[]
candidates=list(range(X_train_ext.shape[1]))
done=False
while not done and len(candidates)>0:
    aux_OLS_lst=[sm.OLS(Y_train,X_train_ext[:,fs_feat+[candidate]]).fit() for candidate in candidates]
    p_vals=[aux_OLS.pvalues[-1] for aux_OLS in aux_OLS_lst]
    p_min=min(p_vals)
    if p_min<p_limit:
        p_min_pos=np.argmin(p_vals)
        print('Adding feature {} with p {:.3f}'.format(candidates[p_min_pos],p_min))
        fs_feat.append(candidates[p_min_pos])
        candidates=candidates[:p_min_pos]+candidates[p_min_pos+1:]
        fs_mult_reg_OLS=aux_OLS_lst[p_min_pos]
    else:
        done=True
fs_mult_reg_OLS.summary()
```
