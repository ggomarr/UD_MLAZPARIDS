# Polynomial regression - X + I(X^2) + ...

## Degree 2

```r
max_degree=2
all_feat_lst=append('Level',paste('I(Level^',2:max_degree,')',sep=''))
```

### Backward elimination

```r
p_limit=0.05
target='Salary'
feat_lst=all_feat_lst
done=FALSE
while(!done & length(feat_lst)>0){
    formula_str=paste(paste(target,'~'),paste(feat_lst,collapse=' + '))
    be_mult_reg=lm(formula=formula_str,df)
    p_vals=summary(be_mult_reg)[4]$coefficients[,4]
    p_max=max(p_vals)
    if(p_max>p_limit & length(feat_lst)>0){
        p_max_feat=names(which.max(p_vals))
        print(noquote(paste('Removing feature',p_max_feat,'with p',format(p_max,digits=3))))
        if(p_max_feat=='(Intercept)'){
            feat_lst=append(c(0),feat_lst)
        } else {    
            feat_lst=feat_lst[feat_lst != p_max_feat]
        }
    } else {
        done=TRUE
    }
}
print(noquote(paste('Final formula:',formula_str)))
summary(be_mult_reg)
```

```r
ggplot() +
    geom_point(aes(x=df$Level,y=df$Salary),
               colour='blue') +
    geom_line(aes(x=df$Level,y=predict(be_mult_reg,df)),
              colour='black') +
    ggtitle('Salary (USD) vs. Level',subtitle='Training data') +
    xlab('Level') + 
    ylab('Salary (USD)')
```

```r
ggplot() +
    geom_point(aes(x=df$Level,y=be_mult_reg$residuals),
               colour='red') +
    ggtitle('Residuals') +
    xlab('Level') + 
    ylab('Salary (USD)')
```

### Forward selection

```r
p_limit=0.05
target='Salary'
feat_lst=c()
feat_candidates=all_feat_lst
done=FALSE
get_reg=function(f,target,f_base,data){
    formula_str=paste(paste(target,'~'),paste(append(f_base,f),collapse=' + '))
    return(lm(formula=formula_str,data))
}
get_marginal_p=function(reg,f_base){
    num_new_features=length(reg$coefficients)-length(f_base)-1
    return(max(tail(summary(reg)[4]$coefficients[,4],num_new_features)))
}
while(!done & length(feat_candidates)>0){
    aux_reg_lst=lapply(feat_candidates,function(f) get_reg(f,target,feat_lst,df))
    p_vals=unlist(lapply(aux_reg_lst,function(r) get_marginal_p(r,feat_lst)))
    p_min=min(p_vals)
    if(p_min<p_limit){
        p_min_pos=which.min(p_vals)
        print(noquote(paste('Adding feature',feat_candidates[p_min_pos],'with p',format(p_min,digits=3))))
        feat_lst=append(feat_lst,feat_candidates[p_min_pos])
        feat_candidates=feat_candidates[-p_min_pos]
        fs_mult_reg=aux_reg_lst[p_min_pos][[1]]
    } else {
        done=TRUE
    }
}
summary(fs_mult_reg)
```

```r
ggplot() +
    geom_point(aes(x=df$Level,y=df$Salary),
               colour='blue') +
    geom_line(aes(x=df$Level,y=predict(fs_mult_reg,df)),
              colour='black') +
    ggtitle('Salary (USD) vs. Level',subtitle='Training data') +
    xlab('Level') + 
    ylab('Salary (USD)')
```

```r
ggplot() +
    geom_point(aes(x=df$Level,y=fs_mult_reg$residuals),
               colour='red') +
    ggtitle('Residuals') +
    xlab('Level') + 
    ylab('Salary (USD)')
```
